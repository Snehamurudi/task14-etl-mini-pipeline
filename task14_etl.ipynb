```python
import os

folders = ["raw", "processed", "output"]
for folder in folders:
    os.makedirs(folder, exist_ok=True)

print("Folders created")
```

```python
import shutil

shutil.move("ecommerce_dataset_updated.csv", "raw/ecommerce_dataset_updated.csv")
```

```python
import pandas as pd

df = pd.read_csv("raw/ecommerce_dataset_updated.csv")
df.head()
```

```python
print("Rows before cleaning:", df.shape[0])
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "")
    .str.replace(")", "")
)
df.columns
```

```python
df["discount_amount"] = df["price_rs."] - df["final_pricers."]
```

```python
df["high_value_purchase"] = df["final_pricers."] > 200
```

```python
customers = df[["user_id"]].drop_duplicates()
```

```python
products = df[["product_id", "category", "price_rs."]].drop_duplicates()
```

```python
orders = df[[
    "user_id",
    "product_id",
    "final_pricers.",
    "payment_method",
    "purchase_date",
    "high_value_purchase"
]]
```

```python
customers.to_csv("output/customers.csv", index=False)
products.to_csv("output/products.csv", index=False)
orders.to_csv("output/orders.csv", index=False)
```

```python
import sqlite3

conn = sqlite3.connect("database.sqlite")

customers.to_sql("customers", conn, if_exists="replace", index=False)
products.to_sql("products", conn, if_exists="replace", index=False)
orders.to_sql("orders", conn, if_exists="replace", index=False)

conn.close()
```
